```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3], [2,5,7])
plt.show()
```

```python
plt.plot([1,2,3], [2,5,7])
plt.title("Matplot Tutorial")
plt.xlabel("This is X")
plt.ylabel("This is Y")
plt.show()
```

```python
x = [2,5,7,10]
y = [12,23,31,45]
x2 = [2,5,7,10]
y2 = [13,18,29,40]

plt.scatter(x,y, label = 'first')
plt.plot(x2,y2, label = 'second')
plt.legend()
plt.show()
```
