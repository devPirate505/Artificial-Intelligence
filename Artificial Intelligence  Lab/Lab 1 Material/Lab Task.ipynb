```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("dataset.csv")

df
```

```python
# df = df.fillna(df.mean())
df = df.dropna()
df
```

```python

G = nx.from_pandas_edgelist(df, 'source', 'target', ['age', 'connections'])

nx.draw(G)
```

```python
num_nodes = G.number_of_nodes()
degrees = dict(G.degree())
print(num_nodes)
print(degrees)
```

```python
options = {
    "font_size": 10,
    "node_size": 500,
    "node_color": "white",
    "edgecolors": "blue",
    "linewidths":4,
    "width": 4,
}

nx.draw_networkx(G, **options)
```

```python
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G)  # positions for all nodes
nx.draw(G, pos, with_labels=True, node_size=700, node_color="lightblue", font_size=10, font_weight="bold")
labels = nx.get_edge_attributes(G, 'connections')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.show()
```

```python
age = df['age']

plt.figure(figsize=(10,6))
plt.hist(age, bins=10, edgecolor='black', alpha = 0.7)
plt.title('Distribution of Age Among Individuals in the Network')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
X = df[['age']].values
y = df[['connections']].values

model = LinearRegression()

model = model.fit(X, y)

# Print the coefficients
# print(f"Intercept: {model.intercept_}")
# print(f"Coefficient: {model.coef_[0]}")

y_pred = model.predict(X)
```

```python
# Plot the data points and the regression line
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, y_pred, color='red', linewidth=2, label='Regression line')
plt.title('Linear Regression: Number of Connections vs. Age')
plt.ylabel('Number of Connections')
plt.xlabel('Age')
plt.legend()
plt.grid(True)
plt.show()
```

```python
new_age = 33

predicted_con = model.predict([[new_age]])

print(f"Predicted number of connections for age {new_age}: {predicted_con[0]}")
# print(predicted_con[0])
```
